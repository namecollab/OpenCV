```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/imagen2.jpg')
im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
face_cascade=cv2.CascadeClassifier('/perro_face.xml')
face_cascade2=cv2.CascadeClassifier('/persona_face.xml')
face_cascade3=cv2.CascadeClassifier('/cat_face.xml')
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
font=cv2.FONT_HERSHEY_SIMPLEX
faces=face_cascade.detectMultiScale(gray,1.5,10)
faces2=face_cascade2.detectMultiScale(gray,1.3,8)
faces3=face_cascade3.detectMultiScale(gray)
```

```python
for(x,y,w,h) in faces:
  points = np.array([[x+round(w/2),y-h],[x+w,y],[x+w,y+h],[x,y+h],[x,y]])
  cv2.polylines(img, np.int32([points]), True, (0, 255, 0), 5)
  cv2.putText(img,'Perro',(x,y),font,0.9,(0,255,0),2)
  
for(x,y,w,h) in faces2:
  points = np.array([[x+round(w/2),y-h],[x+w,y],[x+w,y+h],[x,y+h],[x,y]])
  cv2.polylines(img, np.int32([points]), True, (0, 0, 255), 5)
  cv2.putText(img,'Persona',(x,y),font,0.9,(0,0,255),2)

for(x,y,w,h) in faces3:
  points = np.array([[x+round(w/2),y-h],[x+w,y],[x+w,y+h],[x,y+h],[x,y]])
  cv2.polylines(img, np.int32([points]), True, (255, 0, 0), 5)
  cv2.putText(img,'Gato',(x,y),font,0.9,(255,0,0),2)
```

```python
p,l,m=cv2.split(img)
img=cv2.merge([m,l,p])
plt.imshow(img)
plt.show()
```
