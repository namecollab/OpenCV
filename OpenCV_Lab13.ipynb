```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/imagen1.jpg')
im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
(h, w) = img.shape[:2]
center = (w / 2, h / 2)
angle = 180
scale = 1
M = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img, M, (w, h))
#cv2_imshow(img)
cv2_imshow(rotated)
```

```python
face_cascade=cv2.CascadeClassifier('/perro_face.xml')
face_cascade2=cv2.CascadeClassifier('/persona_face.xml')
face_cascade3=cv2.CascadeClassifier('/cat_face.xml')
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
font=cv2.FONT_HERSHEY_SIMPLEX
faces=face_cascade.detectMultiScale(gray,1.5,10)
faces2=face_cascade2.detectMultiScale(gray,1.5,8)
faces3=face_cascade3.detectMultiScale(gray)
```

```python
for(x,y,w,h) in faces:
	img=cv2.circle(img,(x+round(w/2),y+round(h/2)),round(h/2),(0,255,0),2)
	cv2.putText(img,'Perro',(x,y),font,0.9,(0,255,0),2)

for(z,v,b,n) in faces2:
	img=cv2.circle(img,(z+round(b/2),v+round(n/2)),round(n/2),(0,0,255),2)
	cv2.putText(img,'Persona',(z,v),font,0.9,(0,0,255),2)

for(q,w,e,r) in faces3:
	img=cv2.circle(img,(q+round(e/2),w+round(r/2)),r,(255,0,0),2)
	cv2.putText(img,'Gato',(q,w),font,0.9,(255,0,0),2)
```

```python
p,l,m=cv2.split(img)
img=cv2.merge([m,l,p])
plt.imshow(img)
cv2.imwrite('/cv2-perrogatopersona.jpg',img) 
plt.show()
```
